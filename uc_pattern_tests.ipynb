# HARD
## Study on true positives patterns

```python
hard_kspt_df = pd.read_csv('easy_test_highIou.csv', index_col=0)
hard_kspt_df.cls_conf = hard_kspt_df.cls_conf/hard_kspt_df.obj_conf
hard_kspt_df.sort_values(by='obj_conf', ascending=False)
```

```python
hard_kspt_oc_mean = hard_kspt_df.obj_conf.mean()
hard_kspt_oc_std = hard_kspt_df.obj_conf.std()

hard_kspt_cc_mean = hard_kspt_df.cls_conf.mean()
hard_kspt_cc_std = hard_kspt_df.cls_conf.std()

print(f'obj_conf = {hard_kspt_oc_mean:.2f} +- {hard_kspt_oc_std:.2f}')
print(f'cls_conf = {hard_kspt_cc_mean:.2f} +- {hard_kspt_cc_std:.2f}')
```
